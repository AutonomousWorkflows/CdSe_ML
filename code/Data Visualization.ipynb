```python
import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

# load dataset into Pandas DataFrame
#(cleaned UV data)
df = pd.read_csv('CdSe UV Train Val.csv')
#(cleaned PL data)
df2 = pd.read_csv('CdSe PL Train Val.csv')
```

```python
#Plot scatterplots (colored) of UV FWHM vs Peak, and PL FWHM vs Peak, in both eV and nm)

from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

sb.set(style='ticks')
fig, axs = plt.subplots(2, 1, figsize=(6, 10))
fig.tight_layout(pad=3.0)

sb.scatterplot(data=df, x='Peak_eV', y='FWHM_eV', hue='T_reactor', s=20, alpha  = 0.5, palette = "Spectral_r", ax=axs[0])
sb.scatterplot(data=df2, x='Peak_eV', y='FWHM_eV', hue='T_reactor', s=20, alpha  = 0.5, palette = "Spectral_r", ax=axs[1])

axs[0].set_title('Absorbance Peak (eV)', fontsize=20)
axs[0].set(xlabel=None)
axs[0].set_ylabel('FWHM (eV)', fontsize=16)
axs[0].set_xlim(2,3)
axs[0].set_ylim(0.1, 0.25)
axs[0].legend().set_visible(False)
axs[0].tick_params(axis='both', which='major', labelsize=14)
axs[0].xaxis.set_major_locator(MultipleLocator(0.2))
axs[0].legend(title='Temperature (C)', prop={'size':14},loc='lower right')

axs[1].set_title('Emission Peak (eV)', fontsize=20)
axs[1].set(xlabel=None)
axs[1].set_ylabel('FWHM (eV)', fontsize=16)
axs[1].set_xlabel('Peak (eV)', fontsize=16)
axs[1].set_xlim(2,2.8)
axs[1].set_ylim(0.1, 0.25)
axs[1].legend().set_visible(False)
axs[1].tick_params(axis='both', which='major', labelsize=14)
axs[1].xaxis.set_major_locator(MultipleLocator(0.2))

plt.tight_layout()

#Save figure
#fig.savefig('Scatterplot_Peak_FWHM_eV.svg')
```

```python
#normalize input and output data
min_params = pd.Series([0, 0, 3, 200, 0, 400, 15, 1239.84193/620, 0.05], index =['Cd_mM', 'Se_mM', 'OA_Cd_ratio', 'T_reactor', 't_s','Peak_nm','FWHM_nm','Peak_eV','FWHM_eV'])
max_params = pd.Series([50, 50, 150, 310, 600, 620, 80, 1239.84193/400, 0.4], index =['Cd_mM', 'Se_mM', 'OA_Cd_ratio', 'T_reactor', 't_s','Peak_nm','FWHM_nm','Peak_eV','FWHM_eV'])
df[['Cd_mM', 'Se_mM', 'OA_Cd_ratio', 'T_reactor', 't_s','Peak_nm','FWHM_nm','Peak_eV','FWHM_eV']] = (df[['Cd_mM', 'Se_mM', 'OA_Cd_ratio', 'T_reactor', 't_s','Peak_nm','FWHM_nm','Peak_eV','FWHM_eV']] - min_params) / (max_params- min_params)
df2[['Cd_mM', 'Se_mM', 'OA_Cd_ratio', 'T_reactor', 't_s','Peak_nm','FWHM_nm','Peak_eV','FWHM_eV']] = (df2[['Cd_mM', 'Se_mM', 'OA_Cd_ratio', 'T_reactor', 't_s','Peak_nm','FWHM_nm','Peak_eV','FWHM_eV']] - min_params) / (max_params- min_params)
```

```python
#2D Correlation heatmap on normalized data
df3 = df[['Cd_mM', 'Se_mM', 'OA_Cd_ratio', 'T_reactor', 't_s','Peak_eV','FWHM_eV']]
df3_nm = df[['Cd_mM', 'Se_mM', 'OA_Cd_ratio', 'T_reactor', 't_s','Peak_nm','FWHM_nm']]
df4 = df2[['Cd_mM', 'Se_mM', 'OA_Cd_ratio', 'T_reactor', 't_s','Peak_eV','FWHM_eV']]
df4_nm = df2[['Cd_mM', 'Se_mM', 'OA_Cd_ratio', 'T_reactor', 't_s','Peak_nm','FWHM_nm']]
# creating mask
mask = np.triu(np.ones_like(df3.corr()))
mask[np.triu_indices_from(mask)] = True
mask[np.diag_indices(7)] = False

x_axis_labels = ['Cd conc.','Se conc.','OA to Cd ratio','Reactor temp.','Time (s)','Peak (eV)','FWHM (eV)'] # labels for x-axis
y_axis_labels = ['Cd conc.','Se conc.','OA to Cd ratio','Reactor temp.','Time (s)','Peak (eV)','FWHM (eV)'] # labels for y-axis

x_axis_labels_nm = ['Cd conc.','Se conc.','OA to Cd ratio','Reactor temp.','Time (s)','Peak (nm)','FWHM (nm)'] # labels for x-axis (nm)
y_axis_labels_nm = ['Cd conc.','Se conc.','OA to Cd ratio','Reactor temp.','Time (s)','Peak (nm)','FWHM (nm)'] # labels for y-axis (nm)

fig, axs = plt.subplots(2, 1, figsize=(6, 10))
fig.tight_layout(pad=1)

# plotting a triangle correlation heatmap
sb.heatmap(df3.corr(), fmt='.2f', cmap="PiYG", annot=True, annot_kws={"size": 14}, mask=mask, xticklabels=False, yticklabels=y_axis_labels,cbar = None, vmin=-0.6, vmax=0.6,ax=axs[0])
sb.heatmap(df4.corr(), fmt='.2f', cmap="PiYG", annot=True, annot_kws={"size": 14}, mask=mask, xticklabels=x_axis_labels, yticklabels=y_axis_labels,cbar = None,vmin=-0.6, vmax=0.6,ax=axs[1])

axs[0].set_yticklabels(y_axis_labels, rotation=40, fontsize = 14)
axs[0].set_title('Absorbance (eV)', fontsize=20)
axs[1].set_yticklabels(y_axis_labels, rotation=40, fontsize = 14)

axs[1].set_xticklabels(x_axis_labels, rotation=40, fontsize = 14)
axs[1].set_title('Emission (eV)', fontsize=20)


axs[0].axhline(y=0, color='k',linewidth=3)
axs[0].axhline(y=df3.shape[1], color='k',linewidth=5)
axs[0].axvline(x=0, color='k',linewidth=3)
axs[0].axvline(x=df3.shape[1], color='k',linewidth=5)


axs[1].axhline(y=0, color='k',linewidth=3)
axs[1].axhline(y=df3.shape[1], color='k',linewidth=5)
axs[1].axvline(x=0, color='k',linewidth=3)
axs[1].axvline(x=df3.shape[1], color='k',linewidth=5)


plt.tight_layout()
#Save Figure
#fig.savefig('CorrMap_eV.svg')
```
